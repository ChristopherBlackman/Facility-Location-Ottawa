```python
from main import VoronoiFacilitySelection 
from main import KMeans

import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d


import matplotlib.pyplot as plt
from pprint import pprint
from main import dot_extract
```

```python
v, e = dot_extract(path='./data/test0.dot')
```

```python
model_2 = VoronoiFacilitySelection(random_state=42,max_iter=301,n_cells=5,ret_dia=True,delta=0.5)
loc, diagrams = model_2.fit(v,e)

centroids = np.array([v[l] for l in loc])
verticies = np.array(list(v.values()))
```

```python
loc
```

```python
# Voronoi Diaram FL Step By Step
# Note : Delta is turned up
# Points are moving away, but figure is self adjusting
```

```python
fig = voronoi_plot_2d(diagrams[0])
```

```python
fig = voronoi_plot_2d(diagrams[100])
```

```python
fig = voronoi_plot_2d(diagrams[200])
```

```python
fig = voronoi_plot_2d(diagrams[300])
```

```python
plt.plot(centroids[:,0],centroids[:,1],'o')
plt.plot(verticies[:,0],verticies[:,1],'x')
```

```python
start = []
end = []

for a in e.keys():
    for d in e[a].keys():
        start.append(v[a])
        end.append(v[d])
```

```python
# locations chosen by Voronoi FL
# Blue Dots representing closest intersection
# X marks represenintg optimal in real-data
```

```python
for i in range(len(end)):
    plt.plot([start[i][0],end[i][0]],[start[i][1],end[i][1]],marker = 'o',color='black')
plt.plot(centroids[:,0],centroids[:,1],'o')
plt.plot(diagrams[300].points[:,0],diagrams[300].points[:,1],'x')
```

```python
model_1 = KMeans(random_state=42,n_clusters=5,n_init=10,max_iter=300,ret_dia=True)
loc , dia = model_1.fit(v,e)

centroids = np.array([v[l] for l in loc])
verticies = np.array(list(v.values()))
```

```python
plt.plot(centroids[:,0],centroids[:,1],'o')
plt.plot(verticies[:,0],verticies[:,1],'x')
```
